```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# load datasets

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

### distributions of greenhouse gas score in 2008 and 2018.

```python
gh_08=df_08['greenhouse_gas_score'].plot(kind='hist');
gh_08.set_xlabel('Score')
gh_08.set_title(' 2008 Green House Score');
```

```python
gh_18=df_18['greenhouse_gas_score'].plot(kind='hist');
gh_18.set_xlabel('Score')
gh_18.set_title('2018 Green House Score');
```

### distribution of combined mpg changed from 2008 to 2018?

```python
cmpg_08=df_08['cmb_mpg'].plot(kind='hist')
cmpg_08.set_xlabel('Score')
cmpg_08.set_title('2008 COmbined mpg');
```

```python
cmpg_18=df_18['cmb_mpg'].plot(kind='hist')
cmpg_18.set_xlabel('Score')
cmpg_18.set_title('2008 COmbined mpg');
```
