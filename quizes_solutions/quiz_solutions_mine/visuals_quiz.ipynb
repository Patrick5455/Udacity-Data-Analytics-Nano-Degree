```python
# import and load data
import pandas as pd

%matplotlib inline

df = pd.read_csv('powerplant_data_edited.csv')
df.head()
```

```python
# plot relationship between temperature and electrical output

# scatter plot
df.plot(x='Avg. Temp', y='Energy Output',
        kind='scatter', figsize=(8,8));
#observation there an almosr perfect negative correlation 
# between Energy Output and Avg. Temp
```

```python
# plot distribution of humidity
df['Relative Humidity'].hist(figsize=(8,8));

# observation: there is no normal distribution
```
